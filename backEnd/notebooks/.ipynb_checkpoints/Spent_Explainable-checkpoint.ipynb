```python
import pandas as pd
import numpy as np
import sklearn.ensemble
```

```python
spent = pd.read_csv('../data/spent.csv')
```

```python
spent.describe()
spent.fillna(method='ffill')
```

```python
rf = sklearn.ensemble.RandomForestRegressor(n_estimators=1000)
```

```python
spent_data_headers = ['Occupation',
       'Stay_In_Current_City_Years', 'Marital_Status', 'Product_Category_1',
       'Product_Category_2', 'Product_Category_3']
spent_target_headers = ['Purchase']
spent_data = spent[spent_data_headers]
spent_target = spent[spent_target_headers]
```

```python
train, test, labels_train, labels_test = sklearn.model_selection.train_test_split(spent_data, spent_target, train_size=0.80)
```
